## For comparison, figure out area of polygon traced out by intercostal points

Use shoelace formula [http://en.wikipedia.org/wiki/Shoelace_formula]

```python
def PolygonArea(corners):
    n = len(corners) # of corners
    area = 0.0
    for i in range(n):
        j = (i + 1) % n
        area += corners[i][0] * corners[j][1]
        area -= corners[j][0] * corners[i][1]
    area = abs(area) / 2.0
    return area
```

## Crop out the intercostal areas

```python
crop_width = 300
crop_height = 60
```

```python
format_str = "{}_t{:02d}_crop{}x{}.tif"
```

```python
exclude = [0,1,9,10,11,12,20,21]
```

```python
for root, dirs, files in walk('.'):
    for _file in files:
        if _file.endswith('-AP.tif'):
            patient_id = _file[:-4]
            if patient_id not in intercostal_locations:
                print patient_id, 'not in intercostal_locations' 
            for loc, (x, y) in enumerate( intercostal_locations[ patient_id ] ):
                if loc in exclude:
                    continue
                x0 = x - (crop_width/2)
                y0 = y - (crop_height/2)
                crop_param = '{}x{}+{}+{}'.format(crop_width, crop_height, x0, y0)
                new_name = format_str.format( patient_id, loc, crop_width, crop_height )
                try:
                    #if loc > 10:
                    #    check_output( ['convert', _file, '-crop', crop_param, '-flop', new_name ])
                    #else:
                        check_output( ['convert', _file, '-crop', crop_param, new_name ])
                except:
                    print 'error processing file', _file, ', patient_id' , patient_id
                print new_name
```

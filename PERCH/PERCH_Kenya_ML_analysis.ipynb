```python
import pandas as pd
```

```python
results = pd.DataFrame.from_csv( 'KENYA_CXR_CASE_300x50_RESULTS.tsv', sep='\t')
```

```python
results.head()
```

```python
len(results)
```

```python
correct_results = results[ results['Act. Class'] == results['Pred. Class'] ]
```

```python
len(correct_results)
```

```python
clear_true_positive = correct_results[ correct_results['Act. Class'] == 'CLEAR']
```

```python
len(clear_true_positive)
```

```python
oi_true_positive = correct_results[ correct_results['Act. Class'] == 'OI']
```

```python
len(oi_true_positive)
```

```python
pep_true_positive = correct_results[ correct_results['Act. Class'] == 'PEP']
```

```python
len(pep_true_positive)
```

```python
pep_true_positive['p(PEP)'].mean()
```

```python
pep_true_positive['p(OI)'].mean()
```

```python
pep_true_positive['p(CLEAR)'].mean()
```

```python
pep_true_positive = pep_true_positive[ pep_true_positive['p(PEP)'] > 0.5 ]
```

```python
len(pep_true_positive )
```

```python
oi_true_positive =  oi_true_positive[ oi_true_positive['p(OI)'] > 0.50]
```

```python
len(oi_true_positive)
```

```python
oi_true_positive['p(OI)'].mean()
```

```python
oi_true_positive['p(CLEAR)'].mean()
```

```python
clear_true_positive =  clear_true_positive[ clear_true_positive['p(CLEAR)'] > 0.60]
```

```python
len(clear_true_positive)
```

```python
state_samples = list( pep_true_positive.index ) + list( oi_true_positive.index ) + list( clear_true_positive.index ) 
```

```python
len(state_samples)
```

```python
fs
```

```python
indices = [ fs._contiguous_sample_names.index( name) for name in state_samples]
```

```python
len( indices)
```

```python
sgs = [ fs._contiguous_sample_group_ids[i] for i in indices ]
```

```python
sgs == indices
```

```python
state_fs = fs.SampleReduce( sgs, inplace=False)
```

```python
state_fs.ToFitFile( 'KENYA_CXR_CASE_300x50_STATE3.fit' )
```

```python
state_fs.LDATransform(inplace=True)
```

```python
from wndcharm.FeatureWeights import FisherFeatureWeights
```

```python
fw = FisherFeatureWeights.NewFromFeatureSpace( state_fs ).Threshold()
```

```python
fw
```

```python
fw.Print()
```

```python
state_fs.FeatureReduce(fw, inplace=True)
```

```python
fs
```

```python
rest = fs.SampleReduce(None, sgs, inplace=False)
```

```python
rest.LDATransform( state_fs, inplace=True )
```

```python
rest.FeatureReduce( fw, inplace=True)
```

```python
rest.Normalize( state_fs, inplace=True)
```

```python
from wndcharm.FeatureSpacePrediction import FeatureSpaceClassification
```

```python
result = FeatureSpaceClassification.NewWND5( state_fs, rest, None, quiet=True)
```

```python
result.Print()
```

```python
result.Print()
```

```python
result.Print()
```

```python
all_clear = FeatureSpace.NewFromFitFile( 'KENYA_CXR_CLEAR_ONLY_300x50.fit', long=True)
```

```python
all_clear.FeatureReduce( fw, inplace=True)
```

```python
all_clear.Normalize( state_fs, inplace=True)
```

```python
result = FeatureSpaceClassification.NewWND5( state_fs, all_clear, fw, quiet=True)
```

```python
result.Print()
```
